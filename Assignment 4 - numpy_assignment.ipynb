## 10) Create a 3x3 array with random values. Subtract the row mean from each row.

```python
array2d = np.random.randint(100, size=9).reshape((3,3))
print(f"Actual matrix: \n{array2d}")

first_row_mean = array2d[0].mean()
second_row_mean = array2d[1].mean()
third_row_mean = array2d[2].mean()
print("-"*50)
print(f"means are: {first_row_mean},{second_row_mean},{third_row_mean}")

array2d[0] = array2d[0] - first_row_mean
array2d[1] = array2d[1] - second_row_mean
array2d[2] = array2d[2] - third_row_mean

print("-"*50)
print(f"Edited matrix:\n{array2d}")
```
